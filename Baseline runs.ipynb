```python
from learning import learn_pool

env_config = {'environment_name' : 'Hanabi-Small', 'num_players' : 2, 
              'use_custom_rewards' : False, 'open_hands' : False}

model_config_base = {'scope' : 'agent_test', 'nenvs'  : 32,  'nsteps' : None,
           'fc_input_layers' : [128, ], 'lstm_layers' : [128], 'noisy_fc' : False, 'v_net' : 'copy',
           'gamma' : 0.99, 'ent_coef' : 0.0, 'vf_coef' : 1, 'cliprange' : 0.2,
           'max_grad_norm' : None, 'k' : 24,  
           'lr' : 1e-4, 'lr_half_period' : 50000, 'anneal_lr' : False,
           'normalize_advs': True, 'layer_norm' : False,
           }
rewards_config = {}
folder = './experiments/baselines/'
method = 'self play'
run_copies = 2
```

### checking steps and k

```python
folder = './experiments/baselines/steps_and_k/'
for k in [8, 16]:
    for nsteps in [8, 32]:
        numepochs = 320000 // nsteps
        mc = dict(model_config_base)
        mc['k'] = k
        mc['nsteps'] = nsteps
        runname = 'k:%d_steps:%d' %(k, nsteps)
        learn_pool.run_experiment(runname, [mc], env_config, rewards_config, numepochs, 
                                  folder = folder, method = method, nsteps = nsteps,)
```
