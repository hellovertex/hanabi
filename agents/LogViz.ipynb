```python
import numpy as np
import matplotlib.pyplot as plt
import os

from dopamine.colab import utils
```

```python
stats = utils.load_statistics(os.getcwd()+"/experiments/rainbow_10kit/logs", iteration_number=10000, verbose=True)[0]
```

```python
### Extract avg return per iteration ###

avg_return = [stats['iter%s' % (i)]['average_return'] for i in range(len(stats.values()))]
```

```python
fig = plt.figure(figsize=(20,10))
plt.plot(range(len(stats.keys())),np.ravel(avg_return))
plt.ylabel("Avg. Cards, successfully played")
plt.xlabel("Number of Iterations trained")
plt.title("Rainbow 4Player Self-Play Performance")
fig.show()
```
